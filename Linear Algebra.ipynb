```python
# Coefficient Matrix
a = np.random.randint(1,20,[4,4])
# Dependent variable vector
b = np.array([44,-9,12.0,7])
# solution
x = np.linalg.solve(a,b)
print('Coefficient Matrix')
print(a)
print('\nDependent Variable vector')
print(b)
print('\nSolution')
print(x)
```

```python
#single value decomposition
from skimage import data
# download a sample image
#image = data.coins()
image = data.astronaut()
print(image.dtype, image.min(), image.max(), image.shape)
plt.imshow(image)
plt.show()
```

```python
# normalize image
img = image/255.0
# reorder the axes to have colour channels as the first dimension
img = np.transpose(img, axes=(2,0,1))
```

```python
U,S,V = np.linalg.svd(img)
print(U.shape, S.shape, V.shape)
```

```python
Sigma = np.zeros((3,512,512))
for i in range(3):
    np.fill_diagonal(Sigma[i,:,:], S[i,:])
print(Sigma.shape)
```

```python
reconst = U @ Sigma @ V
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```

```python
k = 50
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```
